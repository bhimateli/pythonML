```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Total Water Consumption (Billion m3)'], bins=30)

plt.title("Distribution of Total Water Consumption")
plt.xlabel("Water Consumption")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(df['Per Capita Water Use (L/Day)'], fill=True)

plt.title("Density Plot - Per Capita Water Use")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(data=df, x='Water Scarcity Level')

plt.title("Water Scarcity Level Distribution")
plt.show()
```

Lambda Functions:
o Use lambda functions for simple data transformations.
o Example: Apply a lambda function to create a new column that
categorizes numerical data into bins.

```python
df['Consumption Category'] = df[
    'Total Water Consumption (Billion m3)'
].apply(lambda x: 'High' if x > 600 else 'Low')
```

```python
df['Consumption Level'] = df[
    'Total Water Consumption (Billion m3)'
].apply(
    lambda x:
        'Low' if x < 400
        else 'Medium' if x < 700
        else 'High'
)
```

```python
df.head()
```

```python
df['Agricultural Fraction'] = df[
    'Agricultural Water Use (%)'
].apply(lambda x: x / 100)
```

User-Defined Functions:
o Write custom functions to perform repetitive tasks or complex
calculations.
o Example: Create a function to calculate the range of salary as low,
medium, high

```python
def consumption_category(value):
    if value < 400:
        return "Low"
    elif value < 700:
        return "Medium"
    else:
        return "High"

consumption_category(1000)
```

```python
def water_stress(row):
    if row['Groundwater Depletion Rate (%)'] > 60:
        return "Severe"
    elif row['Groundwater Depletion Rate (%)'] > 40:
        return "Moderate"
    else:
        return "Low"
df['Stress Category'] = df.apply(water_stress, axis=1)
```

```python
def categorize_levels(value, low, high):
    if value < low:
        return "Low"
    elif value < high:
        return "Medium"
    else:
        return "High"
df['Per Capita Category'] = df[
    'Per Capita Water Use (L/Day)'
].apply(lambda x: categorize_levels(x, 150, 300))

```

Insights and Conclusions

1. Summary of Key Findings

From the exploratory data analysis (EDA), several important insights were observed:

The dataset contains 3900 records with no missing values, indicating good data quality.

Water consumption varies significantly across countries and years.

Agricultural water use represents the largest share of total water consumption compared to industrial and household usage.

Countries with higher per capita water usage generally show higher total water consumption.

Water scarcity levels vary across regions, suggesting differences in water management and environmental conditions.
